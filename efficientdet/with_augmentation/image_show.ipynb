```python
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt
import numpy as np
import cv2

from dataset import TrainDataset
from transform import get_train_transform

annotation = '../../../stratified_kfold/cv_train_1.json'
data_dir = '../../../detection/dataset'
train_dataset = TrainDataset(annotation, data_dir, get_train_transform())

# train_data_loader = DataLoader(
#     train_dataset,
#     batch_size=1,
#     shuffle=False,
#     num_workers=4
# )

image, target, img_id = train_dataset[2]
boxes = target['boxes'].cpu().numpy().astype(np.int32)

numpy_image = image.permute(1,2,0).cpu().numpy()

fig, ax = plt.subplots(1, 1, figsize=(16, 8))

for box in boxes:
    cv2.rectangle(numpy_image, (box[1], box[0]), (box[3],  box[2]), (0, 1, 0), 2)
    
ax.set_axis_off()
ax.imshow(numpy_image);
```

```python
train_data_loader = DataLoader(
    train_dataset,
    batch_size=1,
    shuffle=False,
    num_workers=4
)
```

```python
for i, (image, target, img_id) in enumerate(train_data_loader):
    if i == 1:
        boxes = target['boxes'][0].numpy().astype(np.int32)
        print(boxes)

        # print(image[0])
        numpy_image = cv2.cvtColor(image[0].permute(1,2,0).cpu().numpy(), cv2.COLOR_BGR2RGB)
        print(type(numpy_image), numpy_image.shape)

        fig, ax = plt.subplots(1, 1, figsize=(16, 8))

        for box in boxes:
            print(box)
            cv2.rectangle(img=numpy_image, pt1=(int(box[1]), int(box[0])), pt2=(int(box[3]), int(box[2])), color=(0, 1, 0), thickness=2)

        ax.set_axis_off()
        ax.imshow(numpy_image)
        break
```
